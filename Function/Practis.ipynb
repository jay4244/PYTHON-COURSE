```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

```python
def string(s):
    vowels = 'aeiou'
    if any(ch in vowels for ch in s.lower()):
        print("string was true")
    else:
        print("string is invalid")
    return s

name = input("Enter your name :")
print("string is :", string(name))
```
