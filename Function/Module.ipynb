```python
import random 

print(random.randint(1 , 100))  # 1- 100

print(random.random()) # 0 -  1

l1 =  ["java" , "python","DA", "DS"]

print(random.choice(l1))

print(random.choice("python"))
```

```python
import math


print(math.pi)
print(math.floor(3.15))
print(math.ceil(3.15))
print(math.factorial(5))  # 120
print(math.sqrt(64))
print(math.pow(2,5)) # 2 * 2 * 2 * 2 * 2
```

```python
import datetime


d  = datetime.datetime.now()

print(d)

from datetime import date

d = date.today()
print(d)

d = date (2025 , 12 , 4)

print(d)

d = date (2025 , 12 , 4).day
print(d)

d = date(2025 , 12 , 4).month
print(d)

d = date(2025 ,12 , 4).year

print(d)
```

```python
import datetime 

d = datetime.date.today()

print(d)

print(d.strftime("%A"))  # string format time - day display here

print(d.strftime("%a"))  # THU

print(d.strftime("%B")) # MONTH DISPLAY HERE....

print(d.strftime("%b")) # dec
```

```python
from datetime import date


d1 = date(2025 , 12 , 4)
d2 = date(2025 , 12 , 25)

diff = d2 - d1

print(diff)
```
