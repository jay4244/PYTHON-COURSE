```python
# 1. Write a program to check if a number is positive, negative, or zero.
num = int(input("Enter your number :"))
if num >0:
    print("number is positive")
elif num <0:
    print("number is nagative")
elif num==0:
    print("number is zero")
else:
    print("invalid number")            
```

```python
# 2. Check whether a number is even or odd.
num = int(input("Enter your number :"))
if num%2==0:
    print("number is even",num)
else:
    print("number is odd",num)    
```

```python
# 3. Find the largest among three numbers.
num1 = int(input("Enter your number :"))
num2 = int(input("Enter your number :"))
num3 = int(input("Enter your number :"))
if num1 > num2:
    print(f"{num1} is greater than {num2}")
elif num2 > num3:
    print(f"{num2} is greater than {num3}")
else:
    print(f"{num3} is greater than {num1} and {num2}")        
```

```python
# 4. Check if a year is a leap year.
year = int(input("Enter your number :"))
if year%4==0:
    print(f"{year} is leap year")
else :
    print("year is not a leap year")    
```

```python
# 5. Take marks from the user and print grade (A/B/C/D/F).
marks = int(input("Enter your marks :"))
if marks > 80:
    print(f"student grade is A")
elif marks > 70 and marks < 80:
    print(f"student grade is B")
elif marks > 60 and marks < 70:
    print(f"student grade is C")
elif marks > 40 and marks < 60:
    print(f"student grade is D")            
else:
    print(f"student grade is E")    
```

```python
# 6. Check whether a character is a vowel or consonant.
char = input("Enter your character :")
vowels = 'aeiouAEIOU'
if char in vowels:
    print(f"{char} is vowel")
else:
    print(f"{char} is consonant")
```

```python
# 9. Input a number and print whether it is a single, double, or triple-digit number.
num = int(input("Enter your number :"))
if num >0 and num <10:
    print("number is single digit") 
elif num >=10 and num <100:
    print("number is double digit") 
elif num >=100 and num <1000:
    print("number is triple digit")
else:
    print("number is more than three digit")
```
