```python
l1 = ["java","python","php","datascience"]

print(l1)

l2 = [len(sub) for sub in l1]
print(l2)
```

```python
l1 = ["java","python","php","datascience"]

l2 = [sub.upper() for sub in l1]

print(l1)
print(l2)
```

```python
l1 = ["java","python","php","datascience","c","android"]

l2 = [sub for sub in l1 if len(sub) > 4]

print(l1)
print(l2)
```

```python
l1 = [1,2,3]
l2 = ["java","python","android"]

result = [(i,j) for i in l1 for j in l2]

print(l1)
print(l2)
print(result)
```

```python
shirt_color = ["red","blue","purple"]
bottom = ["jeans","shots"]

result = [(s,b) for s in shirt_color for b in bottom]

print(shirt_color)
print(bottom)
print(result)
```

```python
l1 = ['a','e','i','o','u','b','g','z']
l2 = [char for char in l1 if char in "aeiou"]

print(l1)
print(l2)
```

```python
l1 = ["python","apple","java","banana","uniform"]
l2 = [word for word in l1 if word[0] in "aeiou"]

print(l1)
print(l2)
```
