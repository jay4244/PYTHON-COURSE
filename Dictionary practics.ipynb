```python
# 1. Create a dictionary of 5 students with their marks. Print all keys and values separately.

student = {
    "jay": 78,
    "meet": 67,
    "raj": 45,
    "milan": 52,
    "akash": 37
}

for key,value in student.items():
    print(key,value)
```

```python
# 2. Write a program to check if a key exists in a dictionary.

student = {
    "jay": 78,
    "meet": 67,
    "raj": 45,
    "milan": 52,
    "akash": 37
}

key = input("Enter your key : ")

if key in student:
    print("key Exists in Dictionary.",key)
else:
    print("Dictionary is not Exists.",key)    
```

```python
# 3. Update a value in the dictionary (e.g., change marks of one student).

student = {
    "jay": 78,
    "meet": 67,
    "raj": 45,
    "milan": 52,
    "akash": 37
}

name = input("Enter student name: ")

if name in student:
    new_marks = int(input("Enter new marks: "))
    student[name] = new_marks
    print("Marks updated successfully")
else:
    print("Student not found")

print(student)
```

```python
# 4. Remove a key from the dictionary using `pop()`.

student = {
    "jay": 78,
    "meet": 67,
    "raj": 45,
    "milan": 52,
    "akash": 37
}

key = input("Enter your key : ")
if key in student:
    student.pop(key)
else:
    print("invalid key")
   
print(student)        
```

```python
# 5. Merge two dictionaries into one.

student = {
    "jay": 78,
    "meet": 67,
    "raj": 45,
    "milan": 52,
    "akash": 37
}

teacher = {
    "anjli medam" : "python",
    "ekta medam" : "java",
    "jatin sir" : "fullter",
    "nirav sir" : "english",
    "zalak sir" : ".net", 
}

merged = student | teacher
print(merged)
```
